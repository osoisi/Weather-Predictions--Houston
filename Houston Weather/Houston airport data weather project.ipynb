# Weather Predictions
## Houston Airport Data Weather Project
This project aims to provide accurate weather predictions for the city of Houston, Texas.

## Project Overview
The project uses a Python notebook to read csv data downloaded from the NCDC weather website, preprocess it, and visualize it using Matplotlib and Seaborn. The data is then used to train a machine learning model to predict the weather in Houston. The model is then evaluated using the test data and the results are visualized using Matplotlib and Seaborn.

### Data Source
The data used in this project is sourced from Houston's airport, specifically focusing on key weather variables such as wind speed, pressure, temperature, humidity, and wind direction. This data is crucial for understanding the local weather patterns and making accurate predictions.
Find the data [here](https://www.ncdc.noaa.gov/cdo-web/search)

## Features
* Precipitation - The amount of rain that falls in a day.
* Maximum temperature - The highest temperature recorded in a day.
* Mininmum temperature - The lowest temperature recorded in a day.

```python
def create_predications(predicators, weather_data_ready, reg):
    train = weather_data_ready.loc[:'1999-12-31']
    test = weather_data_ready.loc['2020-01-01':]
    reg.fit(train[predictors], train['Target_max'])
    predictions = reg.predict(test[predictors])
    mean_absolute_error(test['Target_max'], predictions)
    combined = pd.concat([test['Target_max'], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ['actual', 'predictions']
    return error, combined
```

```python
#weather_data_ready['month_max'] = weather_data_ready['Max_temp'].rolling(30).mean()
weather_data_ready.loc[:, 'month_max'] = weather_data_ready['Max_temp'].rolling(30).mean()
```

This code uses .loc to specify that you want to add or modify the 'month_max' column in the weather_data_ready DataFrame. The : indicates that you want to apply this operation to all rows, and 'month_max' specifies the column you're assigning the rolling mean to. This approach makes your intention clear to pandas, helping avoid the SettingWithCopyWarning.
